```python
api = SentinelAPI(copernicus_user, copernicus_password, 'https://scihub.copernicus.eu/dhus')
outdir = '/download_storage/'


footprint = geojson_to_wkt(area_of_interest_geojson)

products = api.query(footprint,
                     date=('20201107', '20201212'),
                     platformname='Sentinel-2',
                     producttype= 'S2MSI2A',
                     cloudcoverpercentage=(0, 90))
"""
date (tuple of (str or datetime) or str, optional) --
A time interval filter based on the Sensing Start Time of the products. Expects a tuple of (start, end), e.g. (“NOW-1DAY”, “NOW”). The timestamps can be either a Python datetime or a string in one of the following formats:

yyyyMMdd
yyyy-MM-ddThh:mm:ss.SSSZ (ISO-8601)
yyyy-MM-ddThh:mm:ssZ
NOW
NOW-<n>DAY(S) (or HOUR(S), MONTH(S), etc.)
NOW+<n>DAY(S)
yyyy-MM-ddThh:mm:ssZ-<n>DAY(S)
NOW/DAY (or HOUR, MONTH etc.) - rounds the value to the given unit
Alternatively, an already fully formatted string such as “[NOW-1DAY TO NOW]” can be used as well.
"""

api.download_all(products, outdir)

```

```python
logging.basicConfig(format='%(asctime)s %(levelname)s %(message)s', level=logging.INFO)
datasets_folder = Path("/indexed_storage/").resolve()
logging.info("Starting indexing process")
for dataset in os.listdir(datasets_folder):
    dataset_folder = Path(os.path.join(datasets_folder, dataset))
    yaml_filename =  str(dataset_folder).split('/')[-1] + ".yaml"
    yaml_full_path = os.path.join(dataset_folder, yaml_filename)
    logging.info("Reading: %s", yaml_full_path)
    subprocess.run(["datacube", "dataset", "add", yaml_full_path])
    logging.info("Indexed dataset %s", dataset_folder)
```
